```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
from scipy import stats

from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from matplotlib import pyplot as plt
```

```python
x=np.linspace(0,1,100)
f=2
```

```python
plt.plot(x,np.sin(2*np.pi*x*f));
```

```python
def pltsin(f=1, a=1):
    plt.figsize=(12,8)
    plt.plot(x,a*np.sin(2*np.pi*x*f))
    plt.ylim(-10,10)
```

```python
pltsin(2,2)
```

```python
interact(pltsin, f=(1,10,0.1), a=(1,10,1));
```

## Normal distribution

```python
x = np.arange(-5.0,5.0,0.1)
yrange = (-0.05,1.05)
def plt_norm(sigma=1, loc=0):
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(111)
    ax1.plot((min(x),max(x)),(0,0), "k:")
    ax1.plot((min(x),max(x)),(1,1), "k:")
    ax1.plot((0,0), yrange, "k:")
    ax1.set_ylim(*yrange)
    ax1.plot(x, stats.norm.pdf(x, loc, sigma), "r-")
    ax2 = ax1.twinx()
    ax2.set_ylim(*yrange)
    ax2.plot(x, stats.norm.cdf(x, loc, sigma), "b-")
```

```python
interact(plt_norm, sigma=(0.01,3,0.1), loc=(-2,2,0.1));
```

## Gamma distribution

```python
x = np.arange(-1.0,10.0,0.1)
yrange = (-0.05,1.05)
def plt_gamma(k=1, loc=0, theta=1):
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(111)
    ax1.plot((min(x),max(x)),(0,0), "k:")
    ax1.plot((min(x),max(x)),(1,1), "k:")
    ax1.plot((0,0), yrange, "k:")
    ax1.set_ylim(*yrange)
    ax1.plot(x, stats.gamma.pdf(x, k, loc, theta), "r-")
    ax2 = ax1.twinx()
    ax2.set_ylim(*yrange)
    ax2.plot(x, stats.gamma.cdf(x, k, loc, theta), "b-")
```

```python
interact(plt_gamma, k=(0.1,10,0.1), loc=(-2,2,0.1), theta=(0,3,0.1));
```

## Gamma + drift

```python
x = np.arange(-1.0,10.0,0.1)
yrange = (-1.05,2.05)
def plt_gammadrift(k=3, loc=0, theta=1, dr_start=0, dr_slope=0):
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(111)
    ax1.plot((min(x),max(x)),(0,0), "k:")
    ax1.plot((min(x),max(x)),(1,1), "k:")
    ax1.plot((0,0), yrange, "k:")
    ax1.set_ylim(*yrange)
    ax1.plot(x, stats.gamma.pdf(x, k, loc, theta), "r-")
    ax2 = ax1.twinx()
    ax2.set_ylim(*yrange)
    ax2.plot(x, dr_start + dr_slope*x + stats.gamma.cdf(x, k, loc, theta), "b-")
```

```python
interact(plt_gammadrift, k=(0.1,10,0.1), loc=(0,10,0.1), theta=(0,3,0.1), dr_start=(-0.5,0.5,0.05), dr_slope=(-0.5,0.5,0.05));
```
